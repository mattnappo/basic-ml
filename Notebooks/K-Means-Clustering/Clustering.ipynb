# Data
It doesn't really matter whether we find some dataset online or if we generate our own. In order to easily change the input data to learn how this algorithm handles certain cases, lets use sklearn to generate our own, highly-customizable dataset.

`sklearn.make_blobs()` will generate a clustered dataset based on certain paramaters.

```python
from sklearn.datasets import make_blobs, make_friedman1

sample_size = 1000
n_clusters = 10

data = make_blobs(
    n_samples = sample_size,
    n_features = 2,
    centers = n_clusters,
    cluster_std = 1,
    random_state = 12
)
```
